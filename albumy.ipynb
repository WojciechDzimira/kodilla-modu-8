# Zadanie: albumy wszechczasów

**Import niezbednych bibliotek**

```python
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
```

**wczytanie danych z linku, uzycie user_agent jest konieczne by strona nie zablokowała połaczenia z botem**

```python
adres_url = 'https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/'
user_agent = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}

data = pd.read_html(adres_url, header=0, storage_options=user_agent)
df = data[0]
```

**AD 1**  
Zamień nagłówki kolumn na polskie odpowiedniki: ['TYTUŁ','ARTYSTA','ROK','MAX POZ']

```python
df.rename(columns={'POS':'POZ', 'TITLE':'TYTUŁ', "ARTIST":'ARTYSTA', "YEAR":"ROK", 'HIGH POSN':'MAX POZ'}, inplace=True)
```

**AD 2**  
Ilu pojedynczych artystów znajduje się na liście?

```python
uniqu_artist_number = df['ARTYSTA'].nunique() 
print(f"Liczba pojedyńczych artystów znajdujacych się na liście top 60 albumów to: {uniqu_artist_number}")
```

**AD 3**  
Które zespoły pojawiają się najczęściej na liście?

```python
top_band = df['ARTYSTA'].value_counts()
print(f"Zespoły które pojawiły się najczęściej na liscie top 60 albumów wszechczasów to: \n{top_band[top_band == top_band.iloc[0]].to_string()}")
```

**AD 4**  
Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.

```python
df.columns = df.columns.str.capitalize()
df.head(2)
```

**AD 5**  
Wyrzuć z tabeli kolumnę ‘Max Poz’


```python
new_df = df.drop(columns=['Max poz'])
new_df.head(2)
```

**AD 6**  
W którym roku wyszło najwięcej albumów znajdujących się na liście?

```python
best_year = df['Rok'].value_counts()
print(f"Lata z największą liczbą albumów w top 60: \n{best_year[best_year == best_year.iloc[0]].to_string()}")
```

**AD 7**  
Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście

```python
between_1960_and_1990 = (df['Rok'] >= 1960) & (df['Rok'] <= 1990)
print(f"Liczba albumów z lat 1960-1990 na liście Top 60 wynosi: {len(df[between_1960_and_1990])}")
```

**AD8**  
W którym roku wydany został najmłodszy album na liście?

```python
print(f"Najmłodszy album na liście top 60 pochodzi z roku: {df['Rok'].max()}")

```

**AD 9**  
Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
sorted_df = df.sort_values(by='Rok', ascending=True)
sorted_df.drop_duplicates(subset=['Artysta'], keep='first', inplace=True)
sorted_df.head(10)
```

**AD 10**
Listę zapisz do pliku csv.

```python
sorted_df.to_csv('earliest_released_albums.csv', index=False)
```
